```python
import networkx as nx
import re
import warnings
warnings.filterwarnings("ignore")

tree = nx.read_graphml("tree.graphml")
# small_tree = nx.read_graphml("small_tree.graphml")
```

```python
root = 'n0'
d = nx.shortest_path_length(tree,root)
```

```python
nodes = list(nx.dfs_postorder_nodes(tree))
pos = {elem: [int(re.findall(r'\d+', elem)[0])*10, d[elem]*10]  for elem in nodes}
```

```python
import matplotlib.pyplot as plt

plt.subplot(121)
plt.figure(figsize=(30, 30))
nx.draw(tree, with_labels=True, font_weight='bold', pos = pos)
```
